***Mean of temp, dewpoint, humidity, wind_speed, pressure***

```python
mean_temp=data.Temp_C.mean()
print(f"Mean temp is:{mean_temp}")
```

```python
mean_windspd=data.Wind Speed_km/h.mean()        
print(f"Mean dew_point is:{mean_windspd}")
```

```python
data.rename(columns={'Wind Speed_km/h':'WindSpeed'},inplace=True)
```

```python
data.head(2)
```

```python
mean_windspd=data.WindSpeed.mean()                  # like this we can find mean of columns
print(f"Mean wind speed is:{mean_windspd}")
```

***Std deviation of wind speed***

```python
std_dev_wind=data.WindSpeed.std()
print(f"Std. deviation of wind_speed is:{std_dev_wind}")
```

***Variance of wind***

```python
var_wind=data.WindSpeed.var()
print(f"Variance of wind speed is:{var_wind}")
```
